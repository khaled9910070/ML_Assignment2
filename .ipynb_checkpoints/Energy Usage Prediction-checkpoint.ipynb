<h3>Re-categorizing the target features</h3>

```python
print( y_total.nunique())
print('Since original dataset has 91 unique values for target_total,\
 it was split into 5 different categories: ')

def generateClass(x):
    if x<30:
        return 5
    elif (x<=50 and x>=30):
        return 4
    elif (x<=70 and x>50):
        return 3
    elif (x<=90 and x>70):
        return 2
    
    else :
        return 1
    
y_total_new_cat = y_total["target_total"].apply(generateClass)
print(y_total_new_cat.unique())
```

<h1>Training Models</h1>

<h3>Logistic Regression</h3>

Not all of the features are equally important. Implemented a function to perform feature selection using RFE.

```python
"""
    Performed feature ranking and returns the number of most important feature
    to be selected to maximise the accurancy.

    Parameters:
    model :  Instace of the model, e.g. LogisticRegression(),LinearRegression(),
    X : The list of independent variables
    y: The list The array of target_function

    >> model = LogisticRegression()
    >> X = []
    >> y = []

    Returns: 
    int: Number of features to be selected to maximise the accurancy.

"""
def get_optimal_num_of_feature(model, X, y):
    accurancies = np.empty(len(X.columns)-1, dtype=object)
    
    for num_feature in range(1,len(X.columns)):
        rfe = RFE(model, num_feature)
        rfe = rfe.fit(X, y.values.ravel())
        
        selected_columns = rfe.support_
        X_columns = X.columns
        X_feature_selected = X.copy()
        
        #Drop the unselected columns
        for index in range(selected_columns.size):
            if (selected_columns[index]==False):
                X_feature_selected.drop(X_columns[index], axis=1, inplace=True)
                
        #TODO: to replace with K-fold validation
        #training and recording the regression
        X_train, X_test, y_train_tot, y_test_tot = train_test_split(X_feature_selected, y, test_size=0.3)
        
        model.fit(X_train, y_train_tot)
        predictions = model.predict(X_train)
        accurancies[num_feature-1] = accuracy_score(y_train_tot, predictions) 

    
    return np.argmax(accurancies)
```

```python
"""
    Return the data frame after feature selection applied to X, chosen nth most important feature.

    Parameters:
    model :  Instace of the model, e.g. LogisticRegression(),LinearRegression(),
    X     : The list of independent variables
    y     : The list The array of target_function
    num_of_feature (int): The number of features to be selected

    Returns: 
    pandas.dataframe: the list of indpenedent variables after feature selection applied.

"""
def get_selected_features(model, X, y, num_of_feature):
    rfe = RFE(model, num_of_feature)
    rfe = rfe.fit(X, y.values.ravel())
    
    selected_columns = rfe.support_
    X_feature_selected = X.copy()

    for index in range(selected_columns.size):
        if (selected_columns[index]==False):
            X_feature_selected.drop(X.columns[index], axis=1, inplace=True)
            
    return X_feature_selected
```

```python
# select the number of feature which provides the best accurancy
lg = LogisticRegression()
number_feature = get_optimal_num_of_feature(lg, X_normalized, y_total_new_cat)
print('number of features: ', number_feature)
df_logistic = get_selected_features(lg, X, y_total_new_cat, number_feature)
```

```python
#Train the model using selected features
X_train, X_test, y_train_tot, y_test_tot = train_test_split(df_logistic, y_total_new_cat, test_size=0.3)

lg.fit(X_train, y_train_tot)
predictions = lg.predict(X_test)
```

```python
print("Accuracy(logistic regression):", accuracy_score(y_test_tot, predictions))
```

The accurancy is ~50%. It's likely becuase the logistic classifier isn't enough to divide the data point. To visualize it, plot the target function with 2 most important features.

```python
df_logistic_important = get_selected_features(lg, X, y_total_new_cat, 2)
```

<h3>Random Forest Classifier</h3>

There are different number of features each of which contributing a little for prediction, it'll be hard to find a single model. Hence, ensemble method is used. Random Forest is used.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#It's not necessary to perform feature selection as most important feature will be selected automatically

X_train, X_test, y_train_tot, y_test_tot = train_test_split(X_normalized, y_total_new_cat, test_size=0.3)

RFC = RandomForestClassifier()
RFC.fit(X_train, y_train_tot)

predictions = RFC.predict(X_test)
print("Accuracy(Random Forest classifier):", accuracy_score(y_test_tot, predictions))
```

```python
# print(confusion_matrix(y_test_tot, predictions))
print(classification_report(y_test_tot, predictions))
```

Random Forest performs better than logistic regression, as multiple decision trees are used for decision making, each tree having different samples and feature (feature bagging)

<h3>Boosting - Adaboost</h3>

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
#feature selection is not necessary as adaboost weigh each feature according to its importance

X_train, X_test, y_train_tot, y_test_tot = train_test_split(X_normalized, y_total_new_cat, test_size=0.3)

ab = AdaBoostClassifier(n_estimators=50)
ab.fit(X_train, y_train_tot)
predictions = ab.predict(X_test)
print("Accuracy(Adaboost classifier):", accuracy_score(y_test_tot, predictions))
```
